# Traer datos directamente del enlace de una pagina web

```python
import requests
import pymysql
import pandas as pd
from io import StringIO
import csv
```

```python
url = 'https://cdn.buenosaires.gob.ar/datosabiertos/datasets/ente-de-turismo/oferta-establecimientos-gastronomicos/oferta_gastronomica.csv'
```

```python
dF = pd.read_csv(url, sep=',')
```

```python
dF.head()
```

```python
dF.info()
```

```python
dF1 = dF.copy()
```

```python
dF1.drop(['long','lat','id','cocina','ambientacion', 'telefono','mail','horario','calle_nombre','calle_altura','calle_cruce','codigo_postal','codigo_postal_argentino'], axis=1, inplace=True)
```

```python
dF1.head()
```

```python
dF1.info()
```

```python
dF1.fillna('NO DATA', inplace=True)
```

```python
dF1.info()
```

```python
dAtos = list(zip(*[dF1[i].values.tolist() for i in dF1])) ## opcion A
```

# FORMA B

```python
archivoA = requests.get(url)
archivoA.encoding = 'utf-8'
archivoA = archivoA.text
```

```python
archivo = csv.reader(StringIO(archivoA))
```

```python
next(archivo) ## salta nombres columnas
dAtos = [(fila[3], fila[4], fila[13], fila[14], fila[15][7:]) for fila in archivo]
```

# CONEXION MYSQL

```python
conexion = pymysql.connect(
    host = 'localhost',
    user = 'root',
    password = 'Mysql2023+',
    database = 'cLase_8'
)
cursorA = conexion.cursor()
```

```python
queryA = 'insert into gastronomica (nombre, categoria, direccion, barrio, comuna) values (%s,%s,%s,%s,%s)'
```

```python
cursorA.executemany(queryA, dAtos)
```

```python
conexion.commit()
```

```python
conexion.close()
```
